```python
import pandas as pd
```

```python
vine_table = pd.read_csv("vine_table.csv")
vine_table.head()
```

```python
total_votes_df = vine_table[vine_table["total_votes"] >= 20]
total_votes_df.head()
```

```python
fif_votes_df = total_votes_df[(total_votes_df["helpful_votes"]/total_votes_df["total_votes"]) >= 0.5]
fif_votes_df.head()
```

```python
ver_y_df = fif_votes_df[fif_votes_df["vine"] == "Y"]
ver_y_df.head()
```

```python
ver_n_df = fif_votes_df[fif_votes_df["vine"] == "N"]
ver_n_df.head()
```

```python
total_paied_reviews = ver_y_df["star_rating"].count()
number_stars_paied = ver_y_df["star_rating"][ver_y_df["star_rating"] == 5].count()
percent_paied_reviews = (number_stars_paied/total_paied_reviews)*100
print(f"total paied reviews:{total_paied_reviews}")
print(f"number of 5 stars for paied vine: {number_stars_paied}")
print(f"percentage of paied 5-star reviews:{percent_paied_reviews:.0f}%")

total_unpaied_reviews = ver_n_df["star_rating"].count()
number_stars_unpaied = ver_n_df["star_rating"][ver_n_df["star_rating"] == 5].count()
percent_unpaied_reviews = (number_stars_unpaied/total_unpaied_reviews)*100
print(f"total unpaied reviews:{total_unpaied_reviews}")
print(f"number of 5 stars for unpaied vine: {number_stars_unpaied}")
print(f"percentage of unpaied 5-star reviews:{percent_unpaied_reviews:.0f}%")
```
